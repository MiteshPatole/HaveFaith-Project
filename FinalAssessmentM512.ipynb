Here, we chose to create a new data which we are going to use for our visualization, we are going to use column to unwrap and columns to combine features, we are going to transform the data by splitting the combined values and organizing them in a well structured format which would later be easier to analyze

```python
def unwrap_combined_data(dataset, column_to_unwrap, columns_to_combine):
    new_data = {}
    for (i,current) in enumerate(dataset[column_to_unwrap]):
        names = [x.strip() for x in current.split(",")]
        for name in names:
            if name == "":
              name = "unknown"
            if column_to_unwrap not in new_data.keys():
              new_data[column_to_unwrap] = []
            new_data[column_to_unwrap].append(name)
            for col_ in columns_to_combine:
              if col_ not in new_data.keys():
                new_data[col_] = []
              new_data[col_].append(dataset[col_][i])
    return pd.DataFrame(new_data)
```

Here, we are using count combined data to count combind frequency of individula values in a specific columns which would make easier to provide insights on the occurence of the value

```python
def count_combined_data(dataset, column_to_unwrap):
    count = dict()
    for current in dataset[column_to_unwrap]:
        names = [x.strip() for x in current.split(",")]
        for name in names:
            if name == "":
              name = "unknown"
            if name not in count:
              count[name] = 0
            count[name] += 1

    counts = {"name": [], "count": []}
    for key, value in count.items():
        counts["name"].append(key)
        counts["count"].append(value)
    return pd.DataFrame(counts)
```

**Q:- How much increase in number of Shows and Movies have the audience experienced in the past years?**

```python
release_year_distribution = df_reclean.groupby(['release_year', 'type']).size().unstack().fillna(0)

fig = go.Figure()

for col in release_year_distribution.columns:
    fig.add_trace(go.Bar(
        x=release_year_distribution.index,
        y=release_year_distribution[col],
        name=col,
        marker=dict(opacity=0.8)
    ))

fig.update_layout(
    barmode='stack',
    title='Release Year Distribution of Movies and TV Shows on Netflix',
    xaxis=dict(title='Release Year'),
    yaxis=dict(title='Count'),
    legend=dict(title='Type')
)

fig.show()
```

**Q:- What is the distribution of Content ratings of Tv show and movies ?**

```python
rating_counts = df_reclean.groupby(['rating', 'type']).size().unstack(fill_value=0)

fig = go.Figure()

for col in rating_counts.columns:
    fig.add_trace(go.Bar(
        x=rating_counts.index,
        y=rating_counts[col],
        name=col,
    ))

fig.update_layout(
    barmode='stack',
    title='Distribution of Content Ratings for Movies and TV Shows on Netflix',
    xaxis=dict(title='Content Rating'),
    yaxis=dict(title='Count'),
    legend=dict(title='Type', orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)

fig.show()
```
